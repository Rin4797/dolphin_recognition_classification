```python
import os
import shutil
from pathlib import Path

dataset_name = '../dolphin_postprocces_dataset'

shutil.rmtree(dataset_name ,ignore_errors=True)
Path(dataset_name).mkdir(parents=True, exist_ok=True)

for class_name in ['0', '1']:
    Path(os.path.join(dataset_name, class_name)).mkdir(parents=True, exist_ok=True)
    

# генерация датасета
for class_name, bboxes in [('1', tp_list), ('0', fp_list)]:
    for i, obj in tqdm(enumerate(bboxes)):
        
        img = cv2.imread(f_names[index2fname[obj[0]]]['path'])
        
        x_min, y_min, x_max, y_max = [int(i) for i in obj[3:]]

        crop_img = img[y_min:y_max, x_min:x_max]
        
        cv2.imwrite(os.path.join(dataset_name, class_name, str(i) + '_' +  index2fname[obj[0]]), crop_img)


fnamedataset2ind = {}
fname2confidence = {}
for class_name, bboxes in [('1', tp_list), ('0', fp_list)]:
    for i, obj in tqdm(enumerate(bboxes)):
        fnamedataset2ind[os.path.join(dataset_name, class_name, str(i) + '_' +  index2fname[obj[0]])] = obj[0]
        fname2confidence[os.path.join(dataset_name, class_name, str(i) + '_' +  index2fname[obj[0]])] = obj[2]
        
ind2fnamedataset = {j:i for i,j in fnamedataset2ind.items()}
```

```python
import torch
from torch.utils.data import Dataset
from torchvision import transforms
from PIL import Image


img_size = 112
size = (img_size, img_size)

class DolphinClassificationDataset(Dataset):

    def __init__(self, dataset: list, transform=None):
        self.img_paths = [i['img_path'] for i in dataset]
        self.dataset = dataset
        self.transform = transform

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):

        img_name = self.dataset[idx]['img_path']
        image = Image.open(img_name)
        
        
        new_size = (img_size, img_size)
        new_im = Image.new("RGB", new_size)
        box = tuple((n - o) // 2 for n, o in zip(new_size, image.size[:2]))
        new_im.paste(image, box)
        

        # shape = image.size[:2]
        label = int(self.dataset[idx]['class'])
        # conf = self.dataset[idx]['confidence']

        if self.transform:
            new_im = self.transform(new_im)

        return new_im, label #, conf, shape[0], shape[1]


train_transform = transforms.Compose([transforms.ToTensor(),
                                      transforms.Resize((img_size, img_size)),
                                      transforms.RandomHorizontalFlip(0.5),
                                      transforms.RandomVerticalFlip(0.5),
                                      transforms.RandomRotation(degrees=30),
                                ])

val_transform = transforms.Compose([transforms.ToTensor(), transforms.Resize((img_size, img_size))
                                ])

dataset = []

for class_name in os.listdir(dataset_name):
    for img_name in os.listdir(os.path.join(dataset_name, class_name)):
        dataset.append({'class': class_name,
                        'img_path': os.path.join(dataset_name, class_name, img_name),
                        'confidence': fname2confidence[os.path.join(dataset_name, class_name, img_name)]})
        
dataset[:2]
```

```python
from sklearn.model_selection import train_test_split
train, val = train_test_split(dataset, test_size=0.3, random_state=42)

train_set = DolphinClassificationDataset(train, train_transform)
val_set = DolphinClassificationDataset(val, val_transform)

print(len(train_set), len(val_set))

batch_size = 64

train_dataloader = torch.utils.data.DataLoader(train_set, batch_size=batch_size, shuffle=True, num_workers=0)
valid_dataloader = torch.utils.data.DataLoader(val_set, batch_size=batch_size, shuffle=False, num_workers=0)
```

```python
import torchvision
from torch import nn

class ResNetDolph(torch.nn.Module):
    def __init__(self, num_classes, resnet_model):
        super(ResNetDolph, self).__init__()
        self.resnet = resnet_model
        self.bn = torch.nn.BatchNorm1d(self.resnet.fc.in_features)
        self.bn2 = torch.nn.BatchNorm1d(128)
        self.fc = torch.nn.Linear(self.resnet.fc.in_features, 128)
        self.fc2 = torch.nn.Linear(128, num_classes)
        self.resnet.fc = nn.Identity()
        
    def forward(self, x):
        x = self.resnet(x)
        # x = torch.hstack([x, torch.vstack([conf, shape_0, shape_1]).T]).to(torch.float32)
        x = self.fc2(self.bn2(nn.ReLU()(self.fc(self.bn(nn.ReLU()(x))))))
        return x

def get_net():
    resnet = torchvision.models.resnet18(weights=torchvision.models.ResNet18_Weights.IMAGENET1K_V1)
    net = ResNetDolph(1, resnet)
    return net
```

```python
from IPython.display import clear_output
from sklearn.metrics import recall_score, precision_score, f1_score, accuracy_score
import matplotlib.pyplot as plt 


def train(net, train_dataloader, valid_dataloader, criterion, optimizer, scheduler=None, epochs=10, device='cpu', checkpoint_epochs=10):
    
    print(f'Training for {epochs} epochs on {device}')
    
    scores = {'train': {'acc': [], 'precision': [], 'recall': [], 'f1': []},
                'val': {'acc': [], 'precision': [], 'recall': [], 'f1': []}}
    
    for epoch in range(1,epochs+1):
        
        
        figure, axis = plt.subplots(2, 2, figsize=(10, 8))
        
        # clear_output(True)
        
        print(f"Epoch {epoch}/{epochs}")
        
        net.train()
        train_loss = torch.tensor(0., device=device)
        # train_accuracy = torch.tensor(0., device=device)
        
        all_y_list_train = []
        all_preds_list_train = []
        
        all_y_list_val = []
        all_preds_list_val = []
        
        for X, y in tqdm(train_dataloader):
            
            X = X.to(device)
            y = y.to(device)
            # conf = conf.to(device)
            # shape_0 = shape_0.to(device)
            # shape_1 = shape_1.to(device)
            
            
            preds = net(X).reshape(-1)
            
            loss = criterion(preds, y.float())
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            with torch.no_grad():
                train_loss += loss * train_dataloader.batch_size
                # train_accuracy += (torch.argmax(preds, dim=1) == y).sum()
                all_y_list_train += y.tolist()
                
                # all_preds_list_train += torch.argmax(preds, dim=1).tolist()
                all_preds_list_train += (torch.nn.Sigmoid()(preds)>0.5).int().tolist()
        
        if valid_dataloader is not None:
            net.eval()
            valid_loss = torch.tensor(0., device=device)
            # valid_accuracy = torch.tensor(0., device=device)
            with torch.no_grad():
                for X, y in valid_dataloader:
                    
                    X = X.to(device)
                    y = y.to(device)
                    # conf = conf.to(device)
                    # shape_0 = shape_0.to(device)
                    # shape_1 = shape_1.to(device)
                    
                    
                    preds = net(X).reshape(-1)
                    
                    loss = criterion(preds, y.float())

                    valid_loss += loss * valid_dataloader.batch_size
                    all_y_list_val += y.tolist()
                    # all_preds_list_val += torch.argmax(preds, dim=1).tolist()
                    all_preds_list_val += (torch.nn.Sigmoid()(preds)>0.5).int().tolist()
                    # valid_accuracy += (torch.argmax(preds, dim=1) == y).sum()
        
        if scheduler is not None: 
            scheduler.step()
            
        scores['train']['acc'].append(accuracy_score(all_y_list_train, all_preds_list_train))
        scores['train']['precision'].append(precision_score(all_y_list_train, all_preds_list_train))
        scores['train']['recall'].append(recall_score(all_y_list_train, all_preds_list_train))
        scores['train']['f1'].append(f1_score(all_y_list_train, all_preds_list_train))
        
        scores['val']['acc'].append(accuracy_score(all_y_list_val, all_preds_list_val))
        scores['val']['precision'].append(precision_score(all_y_list_val, all_preds_list_val))
        scores['val']['recall'].append(recall_score(all_y_list_val, all_preds_list_val))
        scores['val']['f1'].append(f1_score(all_y_list_val, all_preds_list_val))
        
        x = list(range(len(scores['train']['acc'])))
        
        clear_output(True)
                    
        axis[0, 0].plot(x, scores['train']['acc'], label='train')
        axis[0, 0].plot(x, scores['val']['acc'], label='val') 
        axis[0, 0].set_title("Accuracy")
        axis[0, 0].legend()
        
        axis[0, 1].plot(x, scores['train']['precision'], label='train')
        axis[0, 1].plot(x, scores['val']['precision'], label='val')
        axis[0, 1].set_title("Precision") 
        axis[0, 1].legend() 
        
        axis[1, 0].plot(x, scores['train']['recall'], label='train')
        axis[1, 0].plot(x, scores['val']['recall'], label='val') 
        axis[1, 0].set_title("Recall") 
        axis[1, 0].legend()
        
        axis[1, 1].plot(x, scores['train']['f1'], label='train')
        axis[1, 1].plot(x, scores['val']['f1'], label='val')
        axis[1, 1].set_title("F1")
        axis[1, 1].legend()
        
        plt.show()
                
    return net
```

```python
device = 'cuda'

net = get_net().to(device)

criterion = torch.nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(net.parameters(), lr=0.001, weight_decay=0.9)
# optimizer = torch.optim.Adam(net.parameters(), lr=0.0001)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=50, gamma=0.8)

net = train(net, train_dataloader, valid_dataloader, criterion, optimizer, scheduler, 100, device)
```

```python
from PIL.Image import open

def classify_dolphin(detection, model, transform, deivce='cpu'):
  
    img = np.array(open(f_names[index2fname[detection[0]]]['path']))
    x_min, y_min, x_max, y_max = [int(i) for i in detection[3:]]
    crop_img = img[y_min:y_max, x_min:x_max]
        
    new_size = (112, 112)
    new_im = Image.new("RGB", new_size)
    
    box = tuple((n - o) // 2 for n, o in zip(new_size, crop_img.shape[:2]))
    
    new_im.paste(Image.fromarray(crop_img), box)
        
    transformed_img = transform(crop_img)
    
    with torch.no_grad():
      ans = torch.nn.Sigmoid()(model(transformed_img[None, ].to(device))).detach().cpu().numpy()
          
    return ans
```

```python
scores_crops = []

with torch.no_grad():
    for c, bboxes in [('1', tp_list[:]), ('0', fp_list[:])]:
        for bbox in tqdm(bboxes):
                pred = classify_dolphin(bbox, net, val_transform, deivce='cuda')[0]
                if ind2fnamedataset[bbox[0]] in train_set.img_paths:
                    scores_crops.append({'class': int(c), 'pred': pred, 'train': 1})
                else:
                    scores_crops.append({'class': int(c), 'pred': pred, 'train': 0})
```

```python
def get_area_rel(d):
    x_min, y_min, x_max, y_max = [int(i) for i in d[3:]]
    
    area = (x_max - x_min)*(y_max - y_min)
    if (x_max - x_min) >= (y_max - y_min):
        rel = (x_max - x_min) / (y_max - y_min)
    else:
        rel = (y_max - y_min) / (x_max - x_min)
    conf = d[2]
        
    return area, rel, conf
```

```python
# подберем параметры на train, потом глянем качество val
iou_threshold=0.2

subsample_without_gt_train, subsample_without_gt_val = train_test_split(get_predicted_boxes_format({i: results[i] for i in results if i not in (sets_info['train']['names'] + sets_info['val']['names'])}, fname2index), test_size=0.5, shuffle=True, random_state=42)

detections_for_search = get_predicted_boxes_format({i: results[i] for i in results if i in (sets_info['train']['names'])}, fname2index) + subsample_without_gt_train
ground_truths = get_true_boxes_format({i: markup_updated[i] for i in markup_updated if i in (sets_info['train']['names'])}, fname2index)

print_main_metrics(*calc_tp_fp_fn(ground_truths, detections_for_search, iou_threshold))
```

```python
param_search = {}

for i, d in tqdm(enumerate(detections_for_search)):
    area, rel, conf = get_area_rel(d)
    score = classify_dolphin(d, net, val_transform, deivce='cuda')[0]
    param_search[i] = {'area': area, 'rel': rel, 'conf': conf, 'score': score[0]}
```

```python
from itertools import product

df_param_search = pd.DataFrame(param_search).T

scores = {}

for lower_area, upper_area in tqdm(product([min(areas)] + [np.percentile(areas, i) for i in range(5)], [max(areas)] + [np.percentile(areas, i) for i in range(95, 101)])):
    for lower_rel, upper_rel in tqdm(product([min(rels)] + [np.percentile(rels, i) for i in range(5)], [max(rels)] +[np.percentile(rels, i) for i in range(95, 101)])):
        # for score_bound in (np.linspace(0.1, 0.5, 101)):
            indecies = df_param_search[(df_param_search['area'] <= upper_area) & (df_param_search['area'] >= lower_area) & 
                                        (df_param_search['rel'] <= upper_rel) & (df_param_search['rel'] >= lower_rel)].index
            
            post_process_detections = [detections_for_search[i] for i in indecies]
            
            tp, fp, fn = calc_tp_fp_fn(ground_truths, post_process_detections, iou_threshold)
            
            if len(tp) == 0:
                precision = 0
                recall = 0
                f1_score =  0
            else:
                precision = len(tp)/ (len(tp)+ len(fp))
                recall = len(tp)/ (len(tp) + fn)
                f1_score =  2* (precision * recall)/ (precision + recall)
            
            scores[tuple([lower_area, upper_area, lower_rel, upper_rel])] = {'tp': len(tp), 'f1': f1_score, 'recall': recall, 'precision': precision}
            
    # clear_output(True)
```

```python
lower_area, upper_area, lower_rel, upper_rel = pd.DataFrame(scores).T.sort_values('f1', ascending=False)[:50].sort_values('recall', ascending=False).index[0]
```

```python
df_param_search = pd.DataFrame(param_search).T

scores_ = {}

for score_bound in tqdm(np.linspace(0.1, 0.9, 501)):
    indecies = df_param_search[(df_param_search['area'] <= upper_area) & (df_param_search['area'] >= lower_area) & 
                                (df_param_search['rel'] <= upper_rel) & (df_param_search['rel'] >= lower_rel) & 
                                (df_param_search['score'] >= score_bound)].index
    
    # print(indecies)
    
    post_process_detections = [detections_for_search[i] for i in indecies]
    
    tp, fp, fn = calc_tp_fp_fn(ground_truths, post_process_detections, iou_threshold)
    
    if len(tp) == 0:
        precision = 0
        recall = 0
        f1_score =  0
    else:
        precision = len(tp)/ (len(tp)+ len(fp))
        recall = len(tp)/ (len(tp) + fn)
        f1_score =  2* (precision * recall)/ (precision + recall)
    
    scores_[tuple([score_bound])] = {'tp': len(tp), 'f1': f1_score, 'recall': recall, 'precision': precision}
```
